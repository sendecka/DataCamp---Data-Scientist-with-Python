1. Przygotować liniowy wykres 3D, w którym znajdą się wartości z zakresu współrzędnych wyznaczonego pseudolosowo oraz liczebności dobranej pseudolosowo. Ponadto wartości na osiach X, Y, Z zostaną przekształcone przez funkcje trygonometryczne (dostępne w bibliotece NumPy) wyznaczone pseudolosowo. Przykładowo, wartości na osi X zostaną przekształcone funkcją sinus, na osi Y funkcją cosinus, a na osi Z funkcją tangens. W ramach jednej wizualizacji funkcje nie mogą się powtarzać.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Ustawienie ziarna dla powtarzalności generowanych danych pseudolosowych
np.random.seed(0)

# Wygenerowanie pseudolosowych wartości współrzędnych (X, Y, Z) i liczebności
n_points = 100
x = np.random.rand(n_points)  # Losowe wartości od 0 do 1
y = np.random.rand(n_points)  # Losowe wartości od 0 do 1
z = np.random.rand(n_points)  # Losowe wartości od 0 do 1
sizes = np.random.randint(10, 100, size=n_points)  # Losowe rozmiary od 10 do 99

# Przekształcenie współrzędnych przez funkcje trygonometryczne
x_transformed = np.sin(x * 2 * np.pi)  # Sinus
y_transformed = np.cos(y * 2 * np.pi)  # Cosinus
z_transformed = np.tan(z * np.pi - np.pi/2)  # Tangens

# Stworzenie wykresu 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Wykreślenie punktów na wykresie z odpowiednimi przekształconymi współrzędnymi
ax.scatter(x_transformed, y_transformed, z_transformed, s=sizes, alpha=0.6)

# Ustawienie etykiet osi na wykresie
ax.set_xlabel('sin(X)')
ax.set_ylabel('cos(Y)')
ax.set_zlabel('tan(Z)')

# Wyświetlenie wykresu
plt.title('Wykres 3D z przekształconymi współrzędnymi')
plt.show()
```

2. Przygotować punktowy wykres 3D, gdzie znajdą się punkty z zakresu współrzędnych wyznaczonego pseudolosowo oraz liczebności dobranej pseudolosowo. Każdemu punktowi nadać pseudolosowo dobrany kształt (kwadrat, trójkąt, koło) oraz pseudolosowy kolor (zielony, czerwony, niebieski).

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# Ustawienie ziarna dla powtarzalności generowanych danych pseudolosowych
np.random.seed(0)

# Wygenerowanie pseudolosowych wartości współrzędnych (X, Y, Z) i liczebności
n_points = 100
x = np.random.rand(n_points)  # Losowe wartości od 0 do 1
y = np.random.rand(n_points)  # Losowe wartości od 0 do 1
z = np.random.rand(n_points)  # Losowe wartości od 0 do 1
sizes = np.random.randint(50, 200, size=n_points)  # Losowe rozmiary od 50 do 199

# Wybór pseudolosowych kształtów i kolorów dla każdego punktu
shapes = ['o', '^', 's']  # Kształty: koło, trójkąt, kwadrat
colors = ['r', 'g', 'b']  # Kolory: czerwony, zielony, niebieski
shape_indices = np.random.randint(0, len(shapes), size=n_points)
color_indices = np.random.randint(0, len(colors), size=n_points)
selected_shapes = [shapes[idx] for idx in shape_indices]
selected_colors = [colors[idx] for idx in color_indices]

# Stworzenie wykresu 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Wykreślenie punktów na wykresie z odpowiednimi kształtami i kolorami
for i in range(n_points):
    ax.scatter(x[i], y[i], z[i], s=sizes[i], marker=selected_shapes[i], c=selected_colors[i])

# Ustawienie etykiet osi na wykresie
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Dodanie legendy dla kształtów i kolorów
legend_shapes = [plt.Line2D([0], [0], linestyle='None', marker=shapes[i], color='w', markersize=10) for i in range(len(shapes))]
legend_colors = [plt.Line2D([0], [0], linestyle='None', marker='o', color=colors[i], markersize=10) for i in range(len(colors))]
ax.legend(legend_shapes + legend_colors, ['Circle', 'Triangle', 'Square', 'Red', 'Green', 'Blue'])

# Wyświetlenie wykresu
plt.title('Punktowy wykres 3D z różnymi kształtami i kolorami')
plt.show()
```

3. Przygotować wykres powierzchniowy 3D, którego podstawą będzie siatka o zakresie dobranym pseudolosowo. Współrzędnymi powierzchni na osi Z będzie odległość między współrzędnymi znajdującymi się na siatce przetransformowane funkcją trygonometryczną dobraną pseudolosowo.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Ustawienie ziarna dla powtarzalności generowanych danych pseudolosowych
np.random.seed(0)

# Generowanie pseudolosowych danych dla siatki (X, Y) i transformacji (Z)
n_points = 50
x = np.linspace(0, 10, n_points)  # Zakres X od 0 do 10
y = np.linspace(0, 10, n_points)  # Zakres Y od 0 do 10
X, Y = np.meshgrid(x, y)  # Tworzenie siatki (X, Y)

# Wybór pseudolosowej funkcji trygonometrycznej dla przekształcenia (Z)
z_function = np.random.choice([np.sin, np.cos, np.tan])
Z = z_function(X + Y)  # Z = f(X + Y), gdzie f jest losową funkcją trygonometryczną

# Stworzenie wykresu powierzchniowego 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Wykreślenie powierzchni
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Dodanie legendy kolorów
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

# Ustawienie etykiet osi na wykresie
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Ustawienie tytułu wykresu
plt.title('Wykres powierzchniowy 3D z pseudolosową transformacją')

# Wyświetlenie wykresu
plt.show()
```

4. Przygotować "choinkę" o wysokości dobranej pseudolosowo, która będzie zawierała pseudolosowe wykresy punktowe z zadania 2.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Funkcja do generowania pseudolosowych wykresów punktowych
def generate_random_point_plot(ax):
    n_points = 50
    x = np.random.rand(n_points)  # Losowe wartości od 0 do 1
    y = np.random.rand(n_points)  # Losowe wartości od 0 do 1
    z = np.random.rand(n_points)  # Losowe wartości od 0 do 1
    sizes = np.random.randint(10, 100, size=n_points)  # Losowe rozmiary od 10 do 99

    # Losowe wybieranie kształtów (koło, trójkąt, kwadrat) i kolorów (czerwony, zielony, niebieski)
    shapes = ['o', '^', 's']  # Kształty: koło, trójkąt, kwadrat
    colors = ['r', 'g', 'b']  # Kolory: czerwony, zielony, niebieski
    shape_indices = np.random.randint(0, len(shapes), size=n_points)
    color_indices = np.random.randint(0, len(colors), size=n_points)
    selected_shapes = [shapes[idx] for idx in shape_indices]
    selected_colors = [colors[idx] for idx in color_indices]

    # Wykreślenie punktów na wykresie z odpowiednimi kształtami i kolorami
    for i in range(n_points):
        ax.scatter(x[i], y[i], z[i], s=sizes[i], marker=selected_shapes[i], c=selected_colors[i])

# Ustawienie ziarna dla powtarzalności generowanych danych pseudolosowych
np.random.seed(0)

# Generowanie pseudolosowej wysokości choinki
tree_height = np.random.randint(4, 7)  # Losowa wysokość od 4 do 6 warstw choinki

# Stworzenie głównego wykresu
fig = plt.figure(figsize=(10, 10))
ax_main = fig.add_subplot(111, projection='3d')

# Rysowanie "choinki" - warstwa po warstwie
for level in range(tree_height):
    # Ustawienie pozycji dla danej warstwy
    z_position = level * 1.5  # Odległość pomiędzy warstwami

    # Stworzenie subplotu dla warstwy
    ax = fig.add_axes([0.1, 0.1, 0.8, 0.8], projection='3d')
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_zlim(0, 1)
    ax.view_init(elev=10, azim=30)  # Ustawienie perspektywy

    # Wygenerowanie pseudolosowego wykresu punktowego na warstwie
    generate_random_point_plot(ax)

    # "Przesunięcie" wykresu do odpowiedniej wysokości
    for line in ax.get_lines():
        line.set_3d_properties(zs=z_position + line.get_3d_properties()[-1][2])

# Usunięcie osi dla głównego wykresu
ax_main.axis('off')

# Wyświetlenie "choinki"
plt.suptitle('Pseudolosowa choinka z wykresami punktowymi', fontsize=16)
plt.show()
```
