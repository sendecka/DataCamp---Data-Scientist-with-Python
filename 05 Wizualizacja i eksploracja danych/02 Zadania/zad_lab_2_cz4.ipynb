2. Przygotować wykres punktowy dwóch kolumn numerycznych dobranych pseudolosowo z wbudowanego w bibliotekę seaborn zbioru danych wybranego metodą pseudolosową. Podzielić punkty kształtem i kolorem według kolumny, która posiada najmniejszą liczbę unikalnych wartości, lecz nie została przedstawiona na osi x ani na osi y.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import random

# Ustawienie stylu wykresu za pomocą seaborn
sns.set(style="whitegrid")

# Wybór losowego zestawu danych dostępnych w seaborn
available_datasets = sns.get_dataset_names()
chosen_dataset = random.choice(available_datasets)

# Wczytanie wybranego zestawu danych
data = sns.load_dataset(chosen_dataset)

# Wybór dwóch losowych kolumn numerycznych do wykresu punktowego
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns
x_column, y_column = random.sample(list(numerical_columns), 2)

# Znalezienie kolumny z najmniejszą liczbą unikalnych wartości, która nie jest x_column ani y_column
available_columns = [col for col in data.columns if col not in [x_column, y_column]]
unique_value_counts = {col: data[col].nunique() for col in available_columns}
color_shape_column = min(unique_value_counts, key=unique_value_counts.get)

# Stworzenie wykresu punktowego z podziałem punktów według kształtu i koloru
plt.figure(figsize=(10, 6))
sns.scatterplot(x=x_column, y=y_column, hue=color_shape_column, style=color_shape_column, data=data)

# Nadać wizualizacji styl dobierany metodą pseudolosową
styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
chosen_style = random.choice(styles)
sns.set_style(chosen_style)

# Dodanie tytułu i etykiet osi
plt.title(f'Scatter Plot of {x_column} vs {y_column} with {color_shape_column} differentiation')
plt.xlabel(x_column)
plt.ylabel(y_column)

# Wyświetlenie wykresu
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Ustawienie ziarna dla powtarzalności generowanych danych pseudolosowych
np.random.seed(0)

# Wybór pseudolosowych kolumn z danych iris
columns = np.random.choice(['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], size=2, replace=False)

# Załadowanie danych iris z biblioteki Seaborn
iris = sns.load_dataset('iris')

# Wykres punktowy na podstawie wybranych pseudolosowo kolumn
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x=columns[0], y=columns[1], hue=iris.columns[~iris.columns.isin(columns)].min(), style=iris.columns[~iris.columns.isin(columns)].min())
plt.title('Wykres punktowy na podstawie dwóch pseudolosowo wybranych kolumn')
plt.xlabel(columns[0])
plt.ylabel(columns[1])
plt.legend(title=iris.columns[~iris.columns.isin(columns)].min())
plt.show()
```

3. Przygotować wykres liniowy dwóch kolumn numerycznych dobranych pseudolosowo z wbudowanego w bibliotekę seaborn zbioru danych wybranego metodą pseudolosową. Podzielić wizualizację kształtem i kolorem według kolumny, która posiada najmniejszą liczbę unikalnych wartości, lecz nie została przedstawiona na osi x ani na osi y.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Ustawienie ziarna dla powtarzalności generowanych danych pseudolosowych
np.random.seed(0)

# Wybór pseudolosowych kolumn z danych iris
columns = np.random.choice(['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], size=2, replace=False)

# Załadowanie danych iris z biblioteki Seaborn
iris = sns.load_dataset('iris')

# Wybór kolumny do podziału kształtem i kolorem
unique_values_count = iris.select_dtypes(include=['object']).nunique()
division_column = unique_values_count.idxmin()

# Stworzenie wykresu liniowego na podstawie wybranych pseudolosowo kolumn
plt.figure(figsize=(10, 6))
sns.lineplot(data=iris, x=columns[0], y=columns[1], hue=division_column, style=division_column, markers=True, dashes=False)
plt.title(f'Wykres liniowy {columns[1]} od {columns[0]} z podziałem przez {division_column}')
plt.xlabel(columns[0])
plt.ylabel(columns[1])
plt.legend(title=division_column)
plt.show()
```

4. Przygotować wykres słupkowy kolumny dobranej pseudolosowo z wbudowanego w bibliotekę seaborn zbioru danych wybranego metodą pseudolosową. Podzielić słupki kolorem według kolumny, która posiada najmniejszą liczbę unikalnych wartości, lecz nie została przedstawiona na wykresie.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Ustawienie ziarna dla powtarzalności generowanych danych pseudolosowych
np.random.seed(0)

# Wybór pseudolosowej kolumny z danych iris
column = np.random.choice(['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])

# Załadowanie danych iris z biblioteki Seaborn
iris = sns.load_dataset('iris')

# Wybór kolumny do podziału słupków kolorem
unique_values_count = iris.select_dtypes(include=['object']).nunique()
division_column = unique_values_count.idxmin()

# Stworzenie wykresu słupkowego na podstawie wybranej pseudolosowo kolumny
plt.figure(figsize=(8, 6))
sns.countplot(data=iris, x=column, hue=division_column)
plt.title(f'Wykres słupkowy dla kolumny {column} z podziałem przez {division_column}')
plt.xlabel(column)
plt.ylabel('Liczba wystąpień')
plt.legend(title=division_column)
plt.show()
```

5. Przygotować mapę ciepła o liczbie wierszy i kolumn wyznaczonej pseudolosowo. Kolumny nazwać kolejnymi literami alfabetu.

```python
import numpy as np
import matplotlib.pyplot as plt
import string  # Moduł string zawiera listę liter alfabetu

# Ustawienie ziarna dla powtarzalności generowanych danych pseudolosowych
np.random.seed(0)

# Wygenerowanie pseudolosowej liczby wierszy i kolumn dla mapy ciepła
n_rows = np.random.randint(5, 10)  # Losowa liczba wierszy od 5 do 9
n_columns = np.random.randint(5, 10)  # Losowa liczba kolumn od 5 do 9

# Stworzenie pseudolosowych danych do mapy ciepła
data = np.random.rand(n_rows, n_columns)  # Losowe dane w zakresie [0, 1]

# Utworzenie etykiet dla kolumn (nazwanych kolejnymi literami alfabetu)
column_labels = list(string.ascii_uppercase)[:n_columns]  # Lista liter alfabetu

# Wykres mapy ciepła
plt.figure(figsize=(10, 8))
plt.imshow(data, cmap='viridis', aspect='auto', interpolation='nearest')

# Dodanie etykiet dla osi x i y
plt.xticks(np.arange(n_columns), column_labels)
plt.yticks(np.arange(n_rows), np.arange(1, n_rows + 1))

# Dodanie kolorowej skali
plt.colorbar(label='Wartość')

# Dodanie tytułu i etykiet osi
plt.title(f'Mapa ciepła: {n_rows} wierszy x {n_columns} kolumn')
plt.xlabel('Kolumny')
plt.ylabel('Wiersze')

# Wyświetlenie mapy ciepła
plt.show()
```
