# Lab 2c - wizualizacje trójwymiarowe

Wizualizacje danych trójwymiarowych wymagają zastosowania innego podejścia niż klasyczne wizualizacje danych dwuwymiarowych. Każdy punkt nie będzie już definiowany przez dwie, ale przez trzy współrzędne. 



```python
import matplotlib.pyplot as plt  
import numpy as np  
```

## Wykres liniowy

Wykres liniowy (podobnie jak pozostałe wykresy 3D) rozpoczynamy od utworzenia obiektu wykresu (klasa plt.Figure). Następnie, korzystając z metody add_subplot, należy utworzyć obiekt osi wykresu 3D, przekazując wartość '3d' w parametrze projection. W kolejnych krokach należy utworzyć twójwymiarowe współrzędne prezentowanych danych i zaprezentować wizualizację przy użyciu funkcji plot:

```python
fig: plt.Figure = plt.figure(figsize=(10, 7))  
ax: plt.Axes = fig.add_subplot(projection = '3d') 

t: np.ndarray = np.linspace(0, 2 * np.pi, 100)  
x: np.ndarray = np.sin(t) * np.cos(t)  
y: np.ndarray = np.arctan(t) * np.cos(t) 
z: np.ndarray = np.arctan(t)  
ax.plot(x, y, z, label='Proste funkcje trygonometryczne') 

ax.set_xlabel('X')  
ax.set_ylabel('Y')  
ax.set_zlabel('Z')  

ax.legend()  

plt.show()  
```

## Wykres punktowy

Każdy punkt na wykresie oprócz współrzędnych powinien mieć nadany kolor, marker (symbol) oraz rozmiar. Do wizualizacji punktów można wykorzystać funkcję scatter:

```python
fig: plt.Figure = plt.figure(figsize=(10, 7))  
ax: plt.Axes = fig.add_subplot(projection = '3d')  

x: np.ndarray = np.random.randint(0, 30, 30)  
y: np.ndarray = np.random.randint(0, 20, 30)  
z: np.ndarray = np.random.randint(0, 10, 30)  
colors: np.ndarray = np.array(['r'] * 10 + ['g'] * 10 + ['b'] * 30) 

np.random.shuffle(colors)  
markers: np.ndarray = np.array(['s'] * 15 + ['^'] * 15)  
np.random.shuffle(markers)  

for i in range(30):  
  ax.scatter(x[i], y[i], z[i], c=colors[i], marker=markers[i])

ax.set_xlabel('X')  
ax.set_ylabel('Y')  
ax.set_zlabel('Z')  

plt.show()  
```

## Wykres powierzchniowy

Do utworzenia wykresu powierzchniowego służy metoda plot_surface:

```python
from matplotlib import cm  

fig: plt.Figure = plt.figure(figsize=(10, 7))  
ax: plt.Axes = fig.add_subplot(projection = '3d')  

x: np.ndarray = np.arange(-5, 5, 0.25)  
y: np.ndarray = np.arange(-5, 5, 0.25)  

x, y = np.meshgrid(x, y)  

z: np.ndarray = np.sin(np.sqrt(x ** 2 + y ** 2))  


surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm)

ax.set_xlabel('X')  
ax.set_ylabel('Y')  
ax.set_zlabel('Z')  

fig.colorbar(surf, shrink=0.5, aspect=5) 

plt.show()
```

## Wykres słupkowy

Do utworzenia wykresu słupkowego służy metoda bar3d:

```python
fig: plt.Figure = plt.figure(figsize=(10, 7))  
ax: plt.Axes = fig.add_subplot(projection = '3d')  

x: np.ndarray = np.arange(4)  
y: np.ndarray = np.arange(5)  
x, y = np.meshgrid(x, y)  
x, y = x.ravel(), y.ravel()  
z: np.ndarray = x + y  

bottom: np.ndarray = np.zeros_like(z)  

width: int
depth: int
width, depth = 1, 1  

ax.bar3d(x, y, bottom, width, depth, z, shade=True)  

ax.set_xlabel('X')  
ax.set_ylabel('Y')  
ax.set_zlabel('Z')  

ax.set_title('Wykres słupkowy')  

plt.show()
```

## Siatki wykresów trójwymiarowych

Podobnie jak w przypadku wykresów 2D, wykresy 3D można organizować w siatki przy użyciu metody add_subplot wskazując rozmiar siatki oraz pozycję:

```python
fig: plt.Figure = plt.figure(figsize=(20, 5))  

ax: plt.Axes = fig.add_subplot(1, 2, 1, projection='3d')  
t: np.ndarray = np.linspace(0, 2 * np.pi, 100)  
x: np.ndarray = np.sin(t) * np.cos(t)  
y: np.ndarray = np.arctan(t) * np.cos(t) 
z: np.ndarray = np.arctan(t)  
ax.plot(x, y, z, label='Proste funkcje trygonometryczne') 
ax.set_xlabel('X')  
ax.set_ylabel('Y')  
ax.set_zlabel('Z')  
ax.legend()  
ax.set_title('Wykres liniowy')  

ax: plt.Axes = fig.add_subplot(1, 2, 2, projection='3d')  
x: np.ndarray = np.arange(-5, 5, 0.25)  
y: np.ndarray = np.arange(-5, 5, 0.25)  
x, y = np.meshgrid(x, y)  
z: np.ndarray = np.sin(np.sqrt(x ** 2 + y ** 2))  

surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm)
ax.set_xlabel('X')  
ax.set_ylabel('Y')  
ax.set_zlabel('Z')  
fig.colorbar(surf, shrink=0.5, aspect=5)  
ax.set_title('Wykres powierzchniowy')  

fig.suptitle('Dwa wykresy')  

plt.show()
```
