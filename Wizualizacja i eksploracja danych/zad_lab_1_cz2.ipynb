```python
import random

# Wygenerowanie losowych wartości start i end
start = random.randint(0, len(movie_new))
end = random.randint(start, len(movie_new))

# Wybór wierszy z ramki movie_new na podstawie losowych indeksów
movie_random = movie_new.iloc[start:end]
movie_random.head(10)
```
