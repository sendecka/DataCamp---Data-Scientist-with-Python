1. Utworzyć wykres liniowy funkcji 10x + 5 dla dziedziny w zakresie [-10, 60]. Serię danych zwizualizować niebieskimi kwadratami.

```python
import matplotlib.pyplot as plt  

# Tworzenie danych
x = range(-10, 61)
y = [10 * i + 5 for i in x]

# Tworzenie wykresu
plt.plot(x, y, marker='s', color='blue', linestyle='-')

plt.title('Wykres funkcji y = 10x + 5')
plt.xlabel('x')
plt.ylabel('y')

plt.xlim(-10, 60)
plt.ylim(min(y), max(y))
```

2. Utworzyć wykres punktowo-liniowy funkcji liniowej 20x - 4 dla dziedziny w zakresie [0, 72]. Serię danych oznaczyć kolorem czerwonym. Wykres zatytułować "funkcja liniowa". Oznaczyć etykiety osi x oraz y.

```python
# Tworzenie danych
x = range(0, 73)
y = [20 * i - 4 for i in x]

# Tworzenie wykresu punktowo-liniowego
plt.plot(x, y, marker='o', color='red', linestyle='-')

plt.title('Funkcja liniowa')
plt.xlabel('x')
plt.ylabel('y')

plt.xlim(0, 72)
plt.ylim(min(y), max(y))

plt.grid(True)
plt.show()
```

3. Utworzyć wykresy liniowe funkcji liniowych o liczbie wyznaczonej pseudolosowo dla dziedziny w zakresie wyznaczonym pseudolosowo. Współczynniki a oraz b dobrać pseudolosowo. Na wykresie umieścić legendę definiującą każdą z serii danych.

```python
import numpy as np
import matplotlib.pyplot as plt

# Liczba funkcji liniowych do wygenerowania
num_functions = 5

# Tworzenie danych
for i in range(num_functions):
    # Generowanie pseudolosowych współczynników a i b
    a = np.random.uniform(-10, 10)
    b = np.random.uniform(-50, 50)

    # Generowanie dziedziny i wartości funkcji
    x = np.random.uniform(-100, 100, 100)  # Pseudolosowa dziedzina
    y = a * x + b

    # Tworzenie wykresu liniowego
    plt.plot(x, y, label=f'Funkcja {i+1}: y = {a:.2f}x + {b:.2f}')

# Dodanie legendy
plt.legend()

# Dodanie tytułu i etykiet osi
plt.title('Wykresy funkcji liniowych')
plt.xlabel('x')
plt.ylabel('y')

# Wyświetlenie wykresu
plt.grid(True)
plt.show()
```

4. Utworzyć siatkę wykresów dla wykresów funkcji liniowych o liczbie wyznaczonej pseudolosowo dla dziedziny w zakresie wyznaczonym pseudolosowo. Współczynniki a oraz b dobrać pseudolosowo. Na wykresie umieścić legendę definiującą każdą z serii danych.

```python
import numpy as np
import matplotlib.pyplot as plt

# Ustawienia siatki wykresów
num_rows = 3
num_cols = 4

# Tworzenie siatki wykresów
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))

# Iteracja po każdym rzędzie
for i in range(num_rows):
    # Iteracja po każdej kolumnie
    for j in range(num_cols):
        # Generowanie pseudolosowych współczynników a i b
        a = np.random.uniform(-10, 10)
        b = np.random.uniform(-50, 50)
        
        # Generowanie pseudolosowych danych dla dziedziny
        x = np.random.uniform(-20, 20, 50)
        y = a * x + b
        
        # Wykres funkcji liniowej
        axes[i, j].plot(x, y, label=f'y = {a:.2f}x + {b:.2f}')
        
        # Dodanie legendy
        axes[i, j].legend()
        
        # Dodanie tytułu dla każdego wykresu
        axes[i, j].set_title(f'Wykres {i*num_cols + j + 1}')

# Dopasowanie układu
plt.tight_layout()

# Wyświetlenie siatki wykresów
plt.show()
```

5. Utworzyć wykres słupkowy przedstawiający liczbę sprzedanych wycieczek do Sri Lanki, Malediwów, wysp Galapagos, Grecji, wyspy Bali, Hiszpanii i Włoch. Liczbę sprzedanych wycieczek wyznaczyć pseudolosowo. Nadać dowolny tytuł wykresu.

```python
import matplotlib.pyplot as plt
import numpy as np

# Miejsca podróży
destinations = ['Sri Lanka', 'Malediwy', 'Wyspy Galapagos', 'Grecja', 'Wyspy Bali', 'Hiszpania', 'Włochy']

# Liczba sprzedanych wycieczek (pseudolosowa)
sales = np.random.randint(50, 500, size=len(destinations))

# Tworzenie wykresu słupkowego
plt.bar(destinations, sales, color='skyblue')

# Dodanie tytułu i etykiet osi
plt.title('Liczba sprzedanych wycieczek do różnych miejsc')
plt.xlabel('Miejsce podróży')
plt.ylabel('Liczba sprzedanych wycieczek')

# Wyświetlenie wykresu
plt.xticks(rotation=45, ha='right')  # Obrócenie etykiet osi x dla lepszej czytelności
plt.tight_layout()  # Dopasowanie układu
plt.show()
```

6. Utworzyć wykres kołowy liczby projektów według języka programowania. Liczbę projektów i języki dopasować pseudolosowo. Wygenerować legendę.

```python
import matplotlib.pyplot as plt
import numpy as np

# Lista języków programowania
languages = ['Python', 'JavaScript', 'Java', 'C#', 'C++', 'PHP', 'Ruby']

# Pseudolosowa liczba projektów dla każdego języka
projects = np.random.randint(5, 50, size=len(languages))

# Tworzenie wykresu kołowego
plt.figure(figsize=(8, 6))
plt.pie(projects, labels=languages, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab10.colors)

# Dodanie tytułu
plt.title('Liczba projektów według języka programowania')

# Dodanie legendy
plt.legend()

# Wyświetlenie wykresu
plt.axis('equal')  # Zapewnienie, że koło będzie wyglądać jak koło
plt.show()
```

7. Utworzyć siatkę 2 wykresów kołowych:
- pierwszy wykres będzie zawierał udział czterech wybranych marek samochodów w rynku. Liczebności wygenerować pseudolosowo
- drugi wykres będzie zawierał udział czterech wybranych producentów laptopów w rynku. Liczebności wygenerować pseudolosowo.

Dla każdego wykresu wygenerować legendę oraz nadać tytuł.

```python
import matplotlib.pyplot as plt
import numpy as np

# Dane dla samochodów
car_brands = ['Toyota', 'Ford', 'BMW', 'Honda']
car_market_share = np.random.randint(100, 1000, size=len(car_brands))

# Dane dla laptopów
laptop_brands = ['Dell', 'HP', 'Lenovo', 'Apple']
laptop_market_share = np.random.randint(100, 1000, size=len(laptop_brands))

# Tworzenie siatki wykresów
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Pierwszy wykres - samochody
axes[0].pie(car_market_share, labels=car_brands, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab10.colors)
axes[0].set_title('Udział marek samochodów w rynku')
axes[0].legend()

# Drugi wykres - laptopy
axes[1].pie(laptop_market_share, labels=laptop_brands, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab10.colors)
axes[1].set_title('Udział marek laptopów w rynku')
axes[1].legend()

# Wyświetlenie siatki wykresów
plt.tight_layout()
plt.show()
```

8. Utworzyć siatkę pseudolosowych wykresów punktowych oraz ułożyć je "w choinkę". Wygenerować "choinkę" wykresów o pseudolowej wysokości.

```python
import matplotlib.pyplot as plt
import numpy as np

# Ustawienia choinki
rows = 5 
cols = 9 

# Ustawienia figur wykresów
fig, axes = plt.subplots(rows, cols, figsize=(12, 10))

# Iteracja po każdym rzędzie
for i in range(rows):
    # Liczba wykresów w danym rzędzie
    num_plots = i + 1
    
    # Iteracja po każdym wykresie w rzędzie
    for j in range(num_plots):
        # Generowanie pseudolosowych danych
        x = np.random.rand(10)
        y = np.random.rand(10)
        
        # Wykres punktowy
        axes[i, j].scatter(x, y)
        
        # Ukrycie osi
        axes[i, j].axis('off')

# Dopasowanie układu
plt.tight_layout()

# Wyświetlenie choinki wykresów
plt.show()
```
